# List (Storing data in Python)

```python
# creating a list
my_list = [1,3,4,6,7]
# accessing the list element at particular index
print(my_list[3])
# Printing all the list elements
print(my_list)
```

```python
# adding an element in the list
fruits = ["apple","banana","cherry"]
print(fruits)
fruits.append("orange")
print(fruits)
```
